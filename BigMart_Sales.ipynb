**Read CSV file**

```python
import pandas as pd
data = pd.read_csv('/bigmart_data.csv')
```

**Review Dataset**

```python
data.head()
```

```python
pivot = pd.pivot_table(data, index=['Item_Type'], values="Item_Weight", aggfunc= 'mean')
pivot = pivot.rename(columns = {'Item_Weight':'Mean_Weight'})
pivot
```

**Alternate Way of Summarizing Table**

```python
weight_by_item = data.groupby('Item_Type')
weight_by_item.Item_Weight.mean()
```

```python
data.groupby('Item_Type').Item_Weight.mean()
```

```python
pivot_outlet_size = pd.pivot_table(data, index=['Outlet_Size','Outlet_Type'], values='Item_MRP', aggfunc='mean')
pivot_outlet_size
```

```python
pd.crosstab(data['Outlet_Size'],data['Item_Type'], margins= False)
```

```python
average_item_visibility = data.groupby(['Item_Identifier'])['Item_Visibility'].mean().reset_index()
average_item_visibility
```

```python
import numpy as np

data[average_item_visibility_2] = data.groupby(['Item_Identifier'])['Item_Visibility'].transform(np.mean())
```
